```python
## Import needed libraries
import pandas as pd #Dataframes
import numpy as np #Numerics
import json #JSON
import ast #Eval datatype
import gzip ##
from textblob import TextBlob## NPL
import re
```

```python
## Read dataset
df_sen=pd.read_csv("Dataset/Cleaned_Dataset/user_reviews.csv")
```

```python
df_sen["review"][df_sen["review"].isnull()]="Neutral Comment"##Fill NaN values in review collumn with "Neutral Comment" to get Sentiment analysis 1
```

```python
df_sen[df_sen["review"]=="Neutral Comment"] ##View dataframe with column review is "Neutral Comment"
```

```python
## Function to clean column review
def clean_text(text):
    # This code uses a regular expression to remove everything other than text and whitespace
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text

# Aply this function to column review
df_sen['review'] = df_sen['review'].apply(clean_text)
```

```python
df_sen ## view dataset
```

## Sentiment_analysis

### Using textblob library

```python
# FunciÃ³ntion to clasify reviews
def classify_sentiment(polarity):
    if polarity < -0.2:
        return 0 ## Negative
    elif polarity > 0.2:
        return 2 ## Positive
    else:
        return 1 ## Neutral


##Aply this function and create new column "sentiment_analysis"
df_sen['sentiment_analysis'] = df_sen['review'].apply(lambda x: classify_sentiment(TextBlob(x).sentiment.polarity))
```
