## STEAM_GAMES_DATASET ##

#### After fixing dataset structure , next step is to load STEAM_GAMES dataset

```python
filas=[]## Empty list for data
files="Dataset/Original_Dataset/steam_games.json.gz"## Dataset address
with gzip.open(files,"rt",encoding="utf-8") as file:## Open file with gzip compression with utf-8 encoding
    for line in file.readlines(): ## Read line by line
        data=json.loads(line) ##json structure
        filas.append(data) ## Add lines to list
df_sg=pd.DataFrame(filas)## Convert listo to dataframe
```

#### Dataset view

```python
df_sg## view dataset
```

#### Litle description of the dataset

```python
df_sg.info()## Dataset information
```

```python
## Drop rows with NaN values in columns "id"
df_sg.dropna(subset=["id"],inplace=True)
df_sg.reset_index(drop=True,inplace=True)        
```

```python
df_sg## Dataset view
```

```python
df_sg.info()
```

```python
df_sg[df_sg["app_name"].isnull() & df_sg["title"].notnull()]## view app_name column is NaN and title column is not NaN
```

```python
df_sg[df_sg["app_name"].notnull() & df_sg["title"].isnull()]## view title column is NaN and app_name column is not NaN
```

### genres and tags ###

```python
df_sg[df_sg["genres"]!=df_sg["tags"]]## can see genres and tags has same values with different orden and few NaN values
## this could be used to fill NaN values in genres column
```

```python
df_sg[df_sg["genres"].isnull() & df_sg["tags"].notnull()]## view genres column is NaN and tags column is not NaN
```

```python
df_sg[df_sg["tags"].isnull() & df_sg["genres"].notnull()]## view tags column is NaN and genres column is not NaN
```

```python
df_sg["genres"][df_sg["tags"].notnull() & df_sg["genres"].isnull()]=df_sg["tags"]##fill NaN values in genres column with tag values
df_sg["tags"][df_sg["genres"].notnull() & df_sg["tags"].isnull()]=df_sg["genres"]##fill NaN values in tags column with genres values
```

```python
df_sg
```

### release date ###

```python
df_sg["release_date"].value_counts().to_dict() ## view values from release_date column to see their format 
```

```python
df_sg["release_date"][df_sg["release_date"].isnull()]="SinFecha"## Fill NaN values in release_date column
```

```python
## Transform string year column to datetime format
df_sg["year"]=pd.to_datetime(df_sg["release_date"],errors="coerce") ## General
## Special formats
df_sg["year"][df_sg["year"].isnull() & df_sg["release_date"].str.isnumeric()]=pd.to_datetime(df_sg["release_date"][df_sg["year"].isnull() & df_sg["release_date"].str.isnumeric()],format='%Y')
df_sg["year"][df_sg["release_date"]=="Q1 2018"]=pd.to_datetime(df_sg["release_date"][df_sg["release_date"]=="Q1 2018"],format="Q1 %Y")
```

```python
## Extract the year from datetime format
df_sg["year"]=df_sg["year"].dt.year
```

### normalize genres ###

```python
df_sg=df_sg.explode("genres")
```

```python
df_sg["genres"].value_counts() ## view values from column genres
```

### export preliminary_etl_dataset ###

```python
df_sg.to_csv("Dataset/Cleaned_Dataset/steam_games.csv",index=False)
```

## USER_ITEMS_DATASET ##

### next step is to load USER_ITEMS dataset ###

```python
filas=[]## Empty list for data
files="Dataset/Original_Dataset/users_items.json.gz"## Dataset address
with gzip.open(files,"rt",encoding="utf-8") as file:## Open file with gzip compression with utf-8 encoding
    for line in file.readlines():## Read line by line
        data=ast.literal_eval(line)## no json structura and need convert to json format with ast
        filas.append(data)## Add lines to list
df_ui=pd.DataFrame(filas)## Convert list to dataframe
```

### Dataset view

```python
df_ui## Dataset view
```

#### Litle description of the dataset

```python
df_ui.info()## Dataset information
```

### Items column ###

```python
##Normalize data and extract elements in new rows
df_ui=df_ui.explode("items")
df_ui.reset_index(drop=True, inplace=True)
```

```python
df_ui## Dataset view
```

```python
## join dataset with its flattened items column
df_ui = pd.concat([df_ui, pd.json_normalize(df_ui['items'])], axis=1) ## unnested dataframe
```

```python
df_ui## Dataset view
```

### export preliminary_etl_dataset ###

```python
df_ui.to_csv("Dataset/Cleaned_Dataset/user_items.csv",index=False)
```

## USER_REVIWES_DATASET

#### next step is to load USERS_REVIEWS dataset

```python
filas=[]## Empty list for data
files="Dataset/Original_Dataset/user_reviews.json.gz"## Dataset address
with gzip.open(files,"rt",encoding="utf-8") as file:## Open file with gzip compression with utf-8 encoding
    for line in file.readlines():## Read line by line
        data=ast.literal_eval(line)## no json structura and need convert to json format with ast
        filas.append(data)## Add lines to list
df_ur=pd.DataFrame(filas)## Convert list to dataframe
```

#### Dataset view

```python
df_ur## view dataset
```

```python
df_ur.info()
```

### reviews column

```python
##Normalize data and extract elements in new rows
df_ur=df_ur.explode("reviews")
df_ur.reset_index(drop=True, inplace=True)
```

```python
## join dataset with its flattened reviews column
df_ur = pd.concat([df_ur, pd.json_normalize(df_ur['reviews'])], axis=1)
```

```python
df_ur## dataset view
```

### posted column

```python
df_ur["posted"].value_counts().to_dict()## View structure of posted reviews
```

```python
## Extract year of posted
df_ur['year_posted'] = pd.to_datetime(df_ur['posted'], format='Posted %B %d, %Y.', errors='coerce').dt.year
```

### export dataset

```python
df_ur.to_csv("Dataset/Cleaned_Dataset/user_reviews.csv",index=False)
```
